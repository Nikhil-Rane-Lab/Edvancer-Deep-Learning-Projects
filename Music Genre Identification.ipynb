## Importing the required libraries

```python
import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from PIL import Image
import pathlib
import csv


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler


import keras

import warnings
warnings.filterwarnings('ignore')
```

# Extracting the Spectrogram for every Audio

```python
audio_no=0
cmap = plt.get_cmap('inferno')

plt.figure(figsize=(10,10))
genres = 'blues classical country disco hiphop jazz metal pop reggae rock'.split()
for g in genres:
    pathlib.Path(f'img_data/{g}').mkdir(parents=True, exist_ok=True)     
    for filename in os.listdir(f'./output/neat/genres/{g}'):
        songname = f'./output/neat/genres/{g}/{filename}'
        y, sr = librosa.load(songname, mono=True, duration=5)
        print("Working With Audio Number:    ",audio_no)
        plt.specgram(y, NFFT=2048, Fs=2, Fc=0, noverlap=128, cmap=cmap, sides='default', mode='default', scale='dB');
        plt.axis('off');
        plt.savefig(f'img_data/{g}/{filename[:-3].replace(".", "")}.png')
        plt.clf()
        audio_no+=1
```

# Now we will extract relevant features from the Spectrograms.
## We will extract

1. Mel-frequency cepstral coefficients (MFCC)(20 in number)
1. Spectral Centroid,
1. Zero Crossing Rate
1. Chroma Frequencies
1. Spectral Roll-off.

```python
header = 'filename chroma_stft rmse spectral_centroid spectral_bandwidth rolloff zero_crossing_rate'
for i in range(1, 21):
    header += f' mfcc{i}'
header += ' label'
header = header.split()
```

# Also we write the data to a .csv file concurrently

```python
audio_num=0
file = open('data.csv', 'w', newline='')
with file:
    writer = csv.writer(file)
    writer.writerow(header)
genres = 'blues classical country disco hiphop jazz metal pop reggae rock'.split()
for g in genres:
    for filename in os.listdir(f'./output/neat/genres/{g}'):        
        songname = f'./output/neat/genres/{g}/{filename}'
        y, sr = librosa.load(songname, mono=True, duration=30)
        chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
        spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
        spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
        rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
        zcr = librosa.feature.zero_crossing_rate(y)
        mfcc = librosa.feature.mfcc(y=y, sr=sr)
        print("Working for audio number   ",audio_num)
        audio_num+=1
        to_append = f'{filename} {np.mean(chroma_stft)} {np.mean(spec_cent)} {np.mean(spec_bw)} {np.mean(rolloff)} {np.mean(zcr)}'    
        for e in mfcc:
            to_append += f' {np.mean(e)}'
        to_append += f' {g}'
        file = open('data.csv', 'a', newline='')
        with file:
            writer = csv.writer(file)
            writer.writerow(to_append.split())
```

# View with Pandas 

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

# Setting things tidy

```python
data = data.drop(['filename'],axis=1)
data = data.drop(['label'],axis=1)
data=data.rename(columns={"mfcc20": "label"})
```

# Encoding the labels

```python
genre_list = data.iloc[:, -1]
encoder = LabelEncoder()
y = encoder.fit_transform(genre_list)
```

# Scaling the Feature columns

```python
scaler = StandardScaler()
X = scaler.fit_transform(np.array(data.iloc[:, :-1], dtype = float))
```

```python
X[0]
```

## Dividing the data into training and Testing set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

# Building our network

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(256, activation='relu', input_shape=(X_train.shape[1],)))

model.add(layers.Dense(128, activation='relu'))

model.add(layers.Dense(64, activation='relu'))

model.add(layers.Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train,
                    y_train,
                    epochs=50,
                    batch_size=128)
```

```python
test_loss, test_acc = model.evaluate(X_test,y_test)
```

```python
print('test_acc: ',test_acc)
```

## Test accuracy is less than training data accuracy. This indicates Overfitting

# Let's set apart 200 samples in our training data to use as a validation set:

```python
x_val = X_train[:200]   #x_val= less_x_test
partial_x_train = X_train[200:] #less_x_train

y_val = y_train[:200]     #= less_x_test_labels
partial_y_train = y_train[200:] #= less_x_train_labels
```

```python
print(partial_x_train.shape)
```

### Now let's train our network for 20 epochs:

```python
model = models.Sequential()
model.add(layers.Dense(512, activation='relu', input_shape=(X_train.shape[1],)))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(partial_x_train,
          partial_y_train,
          epochs=60,
          batch_size=512,
          validation_data=(x_val, y_val))
```

```python
results = model.evaluate(X_test, y_test)
```
